```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv('Mall_Customers.csv', sep = ',', parse_dates=True)
df.drop(['CustomerID'],axis=1,inplace=True)
df
```

# **1.1. Box plots.**

```python
sns.set_palette("Set2")
sns.boxplot(df['Age'])
plt.title('Box plot of ages of custimers')
```

Выше диаграмма размаха возраста. Мы видим, что в основном, тут люди от 30 до 50 лет

```python
sns.set_palette("Greens")
sns.boxplot(df['Spending Score (1-100)'])
plt.title('Box plot of spending score of customers')
```

Выше диаграмма размаха покупательной способности. Мы видим, что в основном, тут люди, у которых баллы от 35 до 75

# **1.2. Violin plot.**

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.set_palette("cubehelix")
sns.violinplot(x = 'Genre', y = 'Spending Score (1-100)', data = df)
plt.title('You can see relation berween genres and spending scores')
```

Выше мы можем увидеть, как отличается покупательская способноть у мужчин и женщин. Большинтво людей обоих полов набрали около 50 баллов.

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.set_palette("dark")
sns.violinplot(x = 'Genre', y = 'Annual Income (k$)', data = df)
plt.title('You can see relation berween genres and Annual Incomes')
```

Выше мы можем увидеть, как отличаются годовые доходы у мужчин и женщин. Большинтво людей обоих полов получают около 65 тысяч $. Кстати, графики почти идентичны, что говорит о равном материальном положении мужчин и женщин.

# **1.3. Count plots.**

```python
fig, ax = plt.subplots(figsize=(20,5))
sns.countplot(df['Age'])
plt.title('You can see numbers of people of different ages')
```

На графике выше хорошо видно, сколько граждан какого возраста являются клиентами магазина. Больше всего людей в возрасте 32 лет (11 чел), всего 1 человек в возрасте 55 и 1 человек в возрасте 56. Магазин пользуется большой популярностью у молодежи.

```python
plt.figure(figsize=(25,5))
sns.countplot(df['Annual Income (k$)'], palette = 'BrBG')
plt.title('You can see numbers of people who has common annual income')
plt.show()
```

На графике выше хорошо видно, сколько человек получают определенную сумму годового дохода. Самый частый доход - 54 тысячи $ в год и 78 тысяч $ в год.

```python
fig, ax = plt.subplots(figsize=(15,5))
sns.set_palette("BrBG")
sns.countplot(df['Genre'])
plt.title('You can how much men and women go to this shop')
```

На графике выше хорошо видно соотношение количества женщин в мужчин в составе клиентов магазина. У женщин магазин пользуется большей популярностью.

# **1.4. Heatmap.**

```python
sns.heatmap(df.corr(), annot = True, cmap = 'BuPu')
plt.title('Heatmap')
```

```python
sns.heatmap(df.corr(), annot = True, cmap = 'PiYG')
plt.title('Heatmap')
```

# **1.5. Facet grid.**

```python
sns.set_palette('husl')
graph = sns.FacetGrid(df, col ="Genre",  hue ="Genre", height=6, aspect=.75)
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "Spending Score (1-100)", "Annual Income (k$)", edgecolor ="w").add_legend()
# show the object
plt.show()
```

График выше показывает разброс уровней покупательской способности в зависимости с годовым доходам по полу. 

# **1.6. Regplot.**

```python
sns.regplot(x = 'Age', y = 'Annual Income (k$)', data = df, color="xkcd:azure", marker="+")
plt.title('You can see relation between age and annual income')
```

На графике выше хорошо видно границу среднего годового дохода (60 тыс $) и разброс годовых доходов в засисимости от возраста. Наибольшие доходы у лиц в возрасте от 30 до 45 лет.

```python
sns.set_palette("husl")
sns.regplot(x = 'Age', y = 'Spending Score (1-100)', data = df, marker="*")
plt.title('You can see relation between age and spending score')
```

На графике выше хорошо видно общую тенденцию средней покупательсткой способности по возрасту (падает с 60 до 30 с возрастом). Наибольшие баллы у лиц в возрасте от 20 до 40 лет.

# **1.7. lmplot.**

```python
female = df[df['Genre']=='Female']
sns.set_palette("flare")
sns.lmplot('Age', 'Spending Score (1-100)',data=female)
plt.title('You can see relation between age and spending score')
```

# **1.8. Swarm plot**

```python
sns.set_palette("pastel")
sns.swarmplot(x = 'Genre', y = 'Age', data = df)
plt.title('You can see numbers of ladies and gentelmen who goes in this mall')
```

На графике выше хорошо видно разброс людей по полу и возрасту. У женщин равномерное количество представительниц разных возрастов, а у мужчин больше 25-30 летних.

# **1.9. Pair plot**
